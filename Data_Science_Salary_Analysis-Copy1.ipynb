ANALYSIS

```python
# Group by 'employment_type' and calculate mean, median, and count for salaries
# This will help us understand how salaries vary based on different types of employment (Full-Time, Part-Time, Contract, etc.).

salary_by_employment_type = df.groupby('employment_type')['salary'].agg(['mean', 'median', 'count']).reset_index()

# Display the result
# print(salary_by_employment_type)
salary_by_employment_type.round(2)
```

```python
# Melt the DataFrame to have a suitable format for the bar plot
employment_salary_melted = salary_by_employment_type.melt(id_vars='employment_type', value_vars=['mean', 'median'], 
                                                var_name='Salary Type', value_name='Salary')

# Create a bar plot
plt.figure(figsize=(12, 6))
sns.barplot(x='employment_type', y='Salary', hue='Salary Type', data=employment_salary_melted)

# Set the title and labels
plt.title('Mean and Median Salaries by Employment Type', fontsize=16)
plt.xlabel('Employment Type', fontsize=12)
plt.ylabel('Salary (USD)', fontsize=12)

# Show the plot
plt.legend(title='Salary Type')
plt.tight_layout()
plt.show()
```

```python
# Pie chart for mean salary by employment type
employment_type_means = salary_by_employment_type['mean']  # Mean salary for each employment type
employment_labels = salary_by_employment_type['employment_type']  # Employment type labels

plt.figure(figsize=(8, 8))
plt.pie(employment_type_means, labels=employment_labels, autopct='%1.1f%%', startangle=90)
plt.title('Proportion of Mean Salary by Employment Type')
plt.show()
```

KEY INSIGHT:
Contract workers appear to have the potential to earn the most, but there may be significant variation within this group as indicated by the gap between the mean and median.
Full-time employees seem to have a stable salary distribution, with the median being close to the mean.
Freelance and Part-time workers generally earn less, and this might reflect either the nature of part-time work (less commitment) or freelancing not being as lucrative in this context.

```python
# Group by job title and calculate the mean salary, then sort by mean salary in descending order
# find the top 10 job titles with the highest average salaries:
# Interpretation: This will show the top 10 job titles that have the highest average salary.
# You can analyze which roles tend to pay the most.

top_10_highest_paid = df.groupby('job_title')['salary'].mean().sort_values(ascending=False).head(10).reset_index()

# Display the result
top_10_highest_paid.round(2)
```

KEY INSIGHT:
Leadership and specialized engineering roles in data science command the highest salaries, indicating the importance of managerial oversight and technical expertise.
Data science professionals with niche skills (like financial data analysis) or responsibilities tied to key business outcomes (like infrastructure design and management) are highly compensated.
There is a clear stratification in salaries based on job titles, reflecting the varying levels of responsibility, expertise, and demand for different roles within the data science field.
